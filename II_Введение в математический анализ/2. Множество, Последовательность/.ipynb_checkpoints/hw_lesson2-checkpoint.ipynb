### 3. *На языке Python предложить алгоритм вычисляющий численно предел с точностью

```python
from sympy import factorial

def f(n):
    return n/factorial(n)**(1/n)

acc = 0.0000001
i = 2

while f(i) - f(i-1) > acc:
    i = i + 2
    
    
print(f(i))
print(i)
```
